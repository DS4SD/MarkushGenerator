```python
# Runs on mdu kernel 

import os
import glob
import pandas as pd
import ast
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
# Important Note: in s_component_manual_lum_test, the valence is not checked after line 205!
```

```python
dataset = "lum_test"

# Read annotated substituents
substituents = []
for p in glob.glob(os.getcwd() + f"/../../data/{dataset}/stable_annotations/*"):
    annotation = pd.read_csv(p)
    for i, row in annotation.iterrows():
        substituents_with_quotes = ""
        quotes_opened = False
        for s in row[1]:
            if not(quotes_opened):
                if (s == "["):
                    substituents_with_quotes += "['"
                elif (s == "]"):
                    substituents_with_quotes += "']"
                elif (s == ","):
                    substituents_with_quotes += "','"
                elif (s == '"'):
                    quotes_opened = True
                else:
                    substituents_with_quotes += s
            elif (quotes_opened):
                if (s == '"'):
                    quotes_opened = False  
                else:
                    substituents_with_quotes += s
                    
        for substituent in ast.literal_eval(substituents_with_quotes):
            substituents.append(substituent)
```

```python
# Read existing mappings
substituents_mapping = pd.read_csv(os.getcwd() + f"/../../data/text_mappings/s_component_manual_lum_test.csv")
substituents_mapping = list(substituents_mapping["value"].values)
```

```python
# Manually remove integers / integers lists

int_substituents = [
    '0',
    '0-3',
    '0-4',
    '0-5',
    '0-50', 
    '1',
    '1-',
    '1-5',
    '1-50',
    '2',
    '3',
    '4',
    '5',
    '6',
    '0-100',
    '1-100',
    '1-101',
    '1-2',
    '1-3',
    '0-2',
    '4-6',
    '0-15',
    '2-3',
    '3-300',
    '3-4',
    
]
```

```python
substituents = [s for s in np.unique(substituents).tolist() if not(s in int_substituents) and not(s in substituents_mapping)]
for s in substituents:
    
    print('"' + s + '",1')
```

```python
# Display an image for each abbreviation

max_i = 20
for i, s in enumerate(substituents):
    # if s != " N—R1":
    #     continue
    if i > max_i:
        break
    s_found = False
    for p in glob.glob(os.getcwd() + f"/../../data/{dataset}/stable_annotations/*"):
        annotation = pd.read_csv(p)
        for i, row in annotation.iterrows():
            substituents_with_quotes = row.iloc[1][0] + "'" + row.iloc[1][1:-1].replace(",", "','") + "'" + row.iloc[1][-1]
            for substituent in ast.literal_eval(substituents_with_quotes):
                if substituent == s:
                   print(s, p)
                   image_path = os.getcwd() + f"/../../data/{dataset}/images/{p.split('/')[-1][:-4]}.png"
                   image = Image.open(image_path)
                   plt.figure(figsize=(10,10))
                   plt.imshow(image)
                   plt.show()
                   plt.close()
                   s_found = True
                   break
            if s_found:
                break
        if s_found:
            break  
```
